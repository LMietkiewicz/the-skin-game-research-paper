**SHA CHECK**

```python
from collections import Counter
from pathlib import Path

for name, root in {"HAM": Path("datasets/HAM10000"), "BCN": Path("datasets/BCN20000")}.items():
    exts = Counter(p.suffix.lower() for p in root.rglob("*") if p.is_file())
    print(f"{name}: {dict(exts)}")
```

```python
import hashlib
from pathlib import Path
from collections import defaultdict
from tqdm.auto import tqdm

DATASET_DIRS = {
    "HAM": Path("datasets/HAM10000"),
    "BCN": Path("datasets/BCN20000"),
}
IMG_EXTS = {".jpg"}   # verified: HAM and BCN are uniformly .jpg

def sha256_file(path, chunk=1 << 20):
    h = hashlib.sha256()
    with open(path, "rb") as f:
        for block in iter(lambda: f.read(chunk), b""):
            h.update(block)
    return h.hexdigest()

digests = defaultdict(list)
counts = {}
for name, root in DATASET_DIRS.items():
    files = [p for p in root.rglob("*") if p.suffix.lower() in IMG_EXTS]
    counts[name] = len(files)
    for p in tqdm(files, desc=f"hashing {name}"):
        digests[sha256_file(p)].append((name, p.stem, p))

print(f"\nhashed: {counts} | unique digests: {len(digests)}")

dupes = {d: recs for d, recs in digests.items() if len(recs) > 1}
within, cross = defaultdict(list), []
for d, recs in dupes.items():
    names = {r[0] for r in recs}
    (cross if len(names) > 1 else within[next(iter(names))]).append((d, recs))

print(f"duplicate groups: {len(dupes)}  "
      f"(cross-dataset: {len(cross)}, within-HAM: {len(within['HAM'])}, within-BCN: {len(within['BCN'])})")

if cross:
    print("\n--- CROSS-DATASET collisions ---")
    for d, recs in cross:
        print("  " + d[:12] + "  " + ", ".join(f"{ds}:{st}" for ds, st, _ in recs))
else:
    print("\nno cross-dataset collisions — HAM and BCN are byte-disjoint.")

for ds in ("HAM", "BCN"):
    if within[ds]:
        print(f"\n--- within-{ds} duplicate groups ({len(within[ds])}), first 5 ---")
        for d, recs in within[ds][:5]:
            print("  " + d[:12] + "  " + ", ".join(st for _, st, _ in recs))
```
